### Outlier Detection and Trimming

```python
import pandas as pd

stats = df_features.describe()
mean = stats.loc['mean', :]
std = stats.loc['std', :]
lower_bound = mean - 2*std
upper_bound = mean + 2*std
```

```python
columns = list(df_features.columns)
```

```python
df_non_outliers = df_features.copy()
```

```python
count_of_outliers = 0
for index,cl in enumerate(columns):
    for i in range(len(df_non_outliers)):
        if df_non_outliers.iloc[i][cl] >= upper_bound[index]:
            df_non_outliers.iloc[i][cl] = upper_bound[index]
            count_of_outliers += 1
        if df_non_outliers.iloc[i][cl] <= lower_bound[index]:
            df_non_outliers.iloc[i][cl] = lower_bound[index]
            count_of_outliers += 1
```
