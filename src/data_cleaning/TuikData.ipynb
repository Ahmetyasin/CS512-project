# TUIK Dataset contents
This section shows the contents of some of the sections in TUIK (see link below).
* [Link to TUIK data portal](https://data.tuik.gov.tr/)

### Foreign Trade [link](https://data.tuik.gov.tr/Kategori/GetKategori?p=Dis-Ticaret-104)
* Foreign Trade Statistics (Special Trade System):
    * Total imports and exports **(export_import)**
    * Product/Product Groups-Partner Country
    * Partner country/ Country Groups
    * Customs
    * Province and Region
    * Seasonally and Calendar Adjusted External Trade

* Foreign trade indices:
    * Export/Import unit value index **(export/import_unit_value_index)** (based on 2010)
    * Export/Import volume index
    * Terms of trade
    * Export/Import volume index, seasonal-calendar adjusted

* Foreign Trade Statistics Imports of Intermediate Goods
    * Yearly export values per product

* Foreign Trade Statistics (General Trade System)
    * Contains monthly import/export values of specific products

* International Trade in Services Statistics
    * Yearly data about exported and imported services

# Macroeconomic indicators used in different studies

### Thesis
1. Turkish Real Effective Exchange Rate (LEX)
2. USD to Turkish Lira Nominal Exchange Rate (LUSD)
    * Nominal effective exchange rate is the weighted average value of the Turkish lira relative to the basket of the countries’ currencies that have a significant share in Türkiye's foreign trade. Weights are determined using bilateral trade flows. As for real effective exchange rate, it is obtained by purifying relative price effects in nominal effective exchange rate.
3. Turkish Industrial Production (LIND)
4. Inflation (LCPI)
5. Turkish Interbank Lending Rate (LINT)
